# Visualizations:

1. Sales Overview
Objective: Understand the overall sales performance, trends, and patterns over time.


```python
# Convert 'Date' column to datetime
data['Date'] = pd.to_datetime(data['Date'])

min_date = data['Date'].min()
max_date = data['Date'].max()

print(f"Minimum Date: {min_date}")
print(f"Maximum Date: {max_date}")

# Group data by Date to get daily sales
daily_sales = data.groupby('Date')['Amount'].sum()

# Plotting the sales trend over time
plt.figure(figsize=(12, 6))
daily_sales.plot()
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales Amount')
plt.grid(True)
plt.show()
```

2. Product Analysis
Objective: Analyze the distribution of product categories, sizes, and quantities sold.

```python
# Distribution of product categories
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Category')
plt.title('Distribution of Product Categories')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of sizes sold
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Size')
plt.title('Distribution of Product Sizes')
plt.xticks(rotation=45)
plt.show()
```

```python
# Quantity sold by category
plt.figure(figsize=(10, 6))
data.groupby('Category')['Qty'].sum().plot(kind='bar')
plt.title('Total Quantity Sold by Category')
plt.xlabel('Category')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=45)
plt.show()
```

3. Fulfillment Analysis
Objective: Investigate the fulfillment methods used and their effectiveness in delivering orders.

```python
# Count of orders by fulfillment method
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='fulfilled-by')
plt.title('Distribution of Fulfillment Methods')
plt.show()
```

```python
# Analyzing fulfillment method vs. status
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='fulfilled-by', hue='Status')
plt.title('Fulfillment Method vs Order Status')
plt.show()
```

4. Customer Segmentation
Objective: Segment customers based on their buying behavior, location, and other relevant factors.

```python
# For simplicity, segment customers by the frequency of their orders
customer_segments = data.groupby('Order ID')['Amount'].sum()

# Plotting the distribution of customer segments
plt.figure(figsize=(10, 6))
customer_segments.hist(bins=50)
plt.title('Customer Segmentation Based on Total Spending')
plt.xlabel('Total Spending')
plt.ylabel('Number of Customers')
plt.show()
```

5. Geographical Analysis
Objective: Explore the geographical distribution of sales, focusing on states and cities.

```python
# Sales by state
state_sales = data.groupby('ship-state')['Amount'].sum()

plt.figure(figsize=(12, 8))
state_sales.plot(kind='bar')
plt.title('Sales Distribution by State')
plt.xlabel('State')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=90)
plt.show()
```

```python
# Sales by city
city_sales = data.groupby('ship-city')['Amount'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 8))
city_sales.plot(kind='bar')
plt.title('Top 10 Cities by Sales')
plt.xlabel('City')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.show()
```
